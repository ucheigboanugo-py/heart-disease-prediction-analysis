```python
import pandas as pd
from scipy.stats import kendalltau

# Load your dataset (replace 'your_data_file.csv' with the actual path to your dataset)
df = pd.read_csv('C://Users//User//Documents//Projects//Data Analysis//Health//heart.csv')

# Choose correlation coefficient (Kendall's tau)
corr_func = kendalltau

# Calculate correlations for all feature pairs (excluding HeartDisease)
corr_matrix = pd.DataFrame()
for col1 in df.columns[:-1]:
    for col2 in df.columns[:-1]:
        if col1 != col2:
            corr, _ = corr_func(df[col1], df[col2])
            corr_matrix.loc[col1, col2] = corr

# Identify features with strongest correlations (absolute values closest to 1)
corr_matrix_abs = corr_matrix.abs()

# Reshape the correlation matrix for sorting
corr_matrix_flatten = corr_matrix_abs.unstack().sort_values(ascending=False)

# Select unique feature pairs with the strongest correlations
strongest_features = set()
for (feature1, feature2), correlation in corr_matrix_flatten.items():
    if feature1 != feature2 and feature1 not in strongest_features and feature2 not in strongest_features:
        strongest_features.add(feature1)
        strongest_features.add(feature2)

# Print results
print("**Top features most strongly correlated with HeartDisease:**")
for feature in strongest_features:
    print(f"- {feature}: absolute correlation = {corr_matrix_abs[feature].abs().max()}")

# Optional: visualize correlations with heatmap
import matplotlib.pyplot as plt

plt.matshow(corr_matrix_abs)
plt.xticks(range(len(df.columns[:-1])), df.columns[:-1], rotation=45)
plt.yticks(range(len(df.columns[:-1])), df.columns[:-1])
plt.colorbar()
plt.show()
```

```python
import pandas as pd

# Load your data
df = pd.read_csv('C://Users//User//Documents//Projects//Data Analysis//Health//heart.csv')

# Feature Engineering

df_feature_engineered = df.copy()

# 1. Age Groups
df['AgeGroup'] = pd.cut(df['Age'], bins=[20, 30, 40, 50, 60], labels=['20-30', '31-40', '41-50', '51-60'])

# 2. Blood Pressure Categories
df['BloodPressureCategory'] = pd.cut(df['RestingBP'], bins=[0, 120, 130, 140, 180], labels=['Normal', 'Elevated', 'High Stage 1', 'High Stage 2'])

# 3. Interaction Terms
df['Age_MaxHR_Interact'] = df['Age'] * df['MaxHR']

# 4. Average Heart Rate During Exercise
df['AvgExerciseHR'] = df['MaxHR']  # Assuming 'MaxHR' can be considered as a proxy for average exercise heart rate

# 5. Chest Pain Type Encoding
# Assuming ChestPainType is categorical, we can encode it using one-hot encoding
chest_pain_dummies = pd.get_dummies(df['ChestPainType'], prefix='ChestPain')
df = pd.concat([df, chest_pain_dummies], axis=1)

# 6. Fasting Blood Sugar Encoding
# Assuming FastingBS is categorical, we can encode it as well
fasting_bs_dummies = pd.get_dummies(df['FastingBS'], prefix='FastingBS')
df = pd.concat([df, fasting_bs_dummies], axis=1)

# 7. Resting ECG Encoding
# Assuming RestingECG is categorical, we can encode it
resting_ecg_dummies = pd.get_dummies(df['RestingECG'], prefix='RestingECG')
df = pd.concat([df, resting_ecg_dummies], axis=1)

# 8. Exercise Angina Encoding
# Assuming ExerciseAngina is categorical, we can encode it
exercise_angina_dummies = pd.get_dummies(df['ExerciseAngina'], prefix='ExerciseAngina')
df = pd.concat([df, exercise_angina_dummies], axis=1)

# Save the intermediate results
df_feature_engineered.to_csv('C:\\Users\\User\\Desktop\\heart_disease_feature_engineering_intermediate_results1.csv', index=False)

# Display the first few rows of the DataFrame with new features
df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder

# Load your data
df = pd.read_csv("heart_disease_no_outliers.csv")

# Identify categorical columns and encode them
categorical_columns = df.select_dtypes(include=['object']).columns
label_encoder = LabelEncoder()

for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])

# Define features (X) and target variable (y)
features = df.drop('HeartDisease', axis=1)
target = df['HeartDisease']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Initialize the RandomForestClassifier
model = RandomForestClassifier(random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```
